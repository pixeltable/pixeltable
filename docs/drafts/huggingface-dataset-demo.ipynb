```python
import pixeltable as pxt
import datasets
```

```python
# using a small subset of large dataset, can remove 'cr' for for a larger dataset
cohere_wikipedia = datasets.load_dataset("Cohere/wikipedia-2023-11-embed-multilingual-v3",
                                data_dir='cr').select_columns(['url', 'title', 'text', 'emb'])

cohere_wikipedia['train'][0:1]
```

We can easily create a table and import the data from the HF dataset `cl.import_huggingface_dataset`. Pixeltable will map  column types as needed, with the option to over-ride the choice of column types directly via the `schema_override` attribute. In this case, `emb` is an embedding column of dimension 1024, which we provide to the method.

```python
cohere_wikipedia_table = pxt.import_huggingface_dataset('cohere_wikipedia', cohere_wikipedia,
                              schema_override={'emb': pxt.ArrayType((1024,), dtype=pxt.FloatType(), nullable=False)}
                            )
```

The `split` column at the right preserves Huggingface information about whether the data is part of the `train` or `test` split.
